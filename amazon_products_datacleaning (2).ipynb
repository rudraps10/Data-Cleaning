```python
def clean_sales(x):
  x=str(x)
  if 'K' in x:
    num=x.replace('K','').replace('+','').split()[0]
    return int(float(num)*1000)
  elif '+' in x:
    num=x.replace('+','').split()[0]
    return int(num)
  else:
    return np.nan
```

```python
df['sales_volume']=df['sales_volume'].apply(clean_sales)
```

```python
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

def extract_delivery_date_easy(text):
    text = str(text)
    words = text.split()

    for i in range(len(words) - 1):
        if words[i][:3] in months and words[i+1].isdigit():
            return words[i] + ' ' + words[i+1]

    return np.nan
```

```python
df['delivery_date'] = df['delivery'].apply(extract_delivery_date_easy)
df['delivery_date']
```
